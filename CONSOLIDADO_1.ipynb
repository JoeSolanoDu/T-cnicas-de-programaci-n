```python
# OBJETIVO: menú
# Nombre: David Joe Solano Dueñas
# Fecha: 12/04/2025


autenticado = 0
usuario = ""
contrasena_correcta = "77136852"

alumnos_apellidos = ""
alumnos_sexos = ""
alumnos_lentes = ""
alumnos_edades = ""
alumnos_montos = ""
total_alumnos = 0

while True:
    print("""
    BIENVENIDO A MI MENÚ
    1. AUTENTIFICARSE
    2. REGISTRAR DONACIONES
    3. CALCULADORA
    4. REPORTE TOTAL
    5. SALIR DEL PROGRAMA
    """)
    entrada = input("Elija una opción: ")

    if entrada.isdigit():  #método de cadena para conviertir en dígito entero positivo
        menu = int(entrada)
    else:
        print("Por favor, ingrese un número válido.")
        continue

    if menu == 1:
        print("== AUTENTICACIÓN ==")
        usuario = input("Ingrese su nombre: ")
        apellido_paterno = input("Ingrese su apellido paterno: ")
        apellido_materno = input("Ingrese su apellido materno: ")
        contra = input("Ingrese su contraseña (8 dígitos): ")

        if contra == contrasena_correcta:
            print("Autenticación exitosa. ¡Bienvenido", usuario, apellido_paterno, apellido_materno, "!")
            autenticado = 1
        else:
            print("Contraseña incorrecta.")

    elif menu == 2:
        if autenticado != 1:
            print("Debe autenticarse primero (opción 1).")
            continue

        while True:
            seccion = input("Usuario ingrese su sección (A/B/C/D): ").upper()
            if seccion in ("A", "B", "C", "D"):
                break
            print("ERROR, INGRESAR UNA DE LAS OPCIONES (A/B/C/D)")

        print("Su sección es:", seccion)

        cantidad = input("Ingrese la cantidad de alumnos: ")
        if not cantidad.isdigit():
            print("Debe ingresar un número positivo.")
            continue
        n = int(cantidad)

        if n <= 0:
            print("La cantidad debe ser mayor a cero.")
            continue

        contador = 1
        total_donado = 0.0

        alumnos_apellidos = ""
        alumnos_sexos = ""
        alumnos_lentes = ""
        alumnos_edades = ""
        alumnos_montos = ""
        total_alumnos = n #Fixed: Indentation adjusted to be within the elif block

        while contador <= n:
            print("\n=== ALUMNO", contador, "===")
            apellido = input("Ingrese el apellido: ")

            sexo = ""
            while sexo not in ("M", "F"):
                sexo = input("Ingrese el sexo (M/F): ").upper()

            lentes = ""
            while lentes not in ("1", "0"):
                lentes = input("¿Usa lentes? (1=Sí / 0=No): ")

            edad_txt = input("Ingrese la edad (13 a 18): ")
            if edad_txt.isdigit():
                edad = int(edad_txt)
                if edad < 13 or edad > 18:
                    print("Edad fuera de rango.")
                    continue
            else:
                print("Debe ser un número válido.")
                continue

            monto_txt = input("Ingrese el monto donado: ")
            try_monto = monto_txt.replace(".", "")
            if try_monto.isdigit():
                monto = float(monto_txt)
                if monto <= 0:
                    print("El monto debe ser positivo.")
                    continue
            else:
                print("Monto inválido.")
                continue

            total_donado += monto

            alumnos_apellidos += apellido + "|"
            alumnos_sexos += sexo + "|"
            alumnos_lentes += lentes + "|"
            alumnos_edades += str(edad) + "|"
            alumnos_montos += str(monto) + "|"

            print("\n=== REPORTE DEL ALUMNO", contador, "===")
            print("Apellido:", apellido)
            print("Sexo:", "Masculino" if sexo == "M" else "Femenino")
            print("Usa lentes:", "Sí" if lentes == "1" else "No")
            print("Edad:", edad)
            print("Monto donado:", monto)
            print("===========================\n")

            contador += 1

    elif menu == 3:
        while True:
            print("""
Bienvenido a mi submenú
1. Suma (2 números)
2. Resta (2 números)
3. Multiplicación (2 números)
4. División (2 números)
5. Módulo de (2 números)
6. Volver al menú principal
            """)
            sub = input("Elija una opción: ")

            if sub == "6":
                break

            if sub in ("1", "2", "3", "5"):
                a = float(input("Ingrese el primer número (positivo): "))
                b = float(input("Ingrese el segundo número (positivo): "))
                if a <= 0 or b <= 0:
                    print("Ambos deben ser positivos.")
                    continue

            if sub == "1":
                print("La suma es:", a + b)
            elif sub == "2":
                print("La resta es:", a - b)
            elif sub == "3":
                print("La multiplicación es:", a * b)
            elif sub == "4":
                a = float(input("Ingrese el primer número: "))
                b = float(input("Ingrese el segundo número (distinto de cero): "))
                if b == 0:
                    print("No se puede dividir por cero.")
                    continue
                print("La división es:", a / b)
            elif sub == "5":
                print("El módulo es:", a % b)
            else:
                print("Opción no válida.")

    elif menu == 4:
        if autenticado == 0:
            print("Debe autenticarse primero (opción 1).")
            continue

        print("\n=== REPORTE TOTAL ===")
        if alumnos_apellidos == "":
            print("No hay alumnos registrados.")
        else:
            total_donado = 0.0
            total_m = 0
            total_f = 0
            lentes_m = 0
            lentes_f = 0
            sin_lentes = 0
            menores = 0
            mayores = 0

            i = 0
            while i < total_alumnos:
                print("\nAlumno")
                print("Apellido:", alumnos_apellidos.split("|")[i])
                sexo_actual = alumnos_sexos.split("|")[i]
                print("Sexo:", "Masculino" if sexo_actual == "M" else "Femenino")
                lentes_actual = alumnos_lentes.split("|")[i]
                print("Usa lentes:", "Sí" if lentes_actual == "1" else "No")
                edad_actual = int(alumnos_edades.split("|")[i])
                print("Edad:", edad_actual)
                monto_actual = float(alumnos_montos.split("|")[i])
                print("Monto donado:", monto_actual)
                total_donado += monto_actual

                if sexo_actual == "M":
                    total_m += 1
                    if lentes_actual == "1":
                        lentes_m += 1
                else:
                    total_f += 1
                    if lentes_actual == "1":
                        lentes_f += 1

                if lentes_actual == "0":
                    sin_lentes += 1

                if edad_actual < 18:
                    menores += 1
                else:
                    mayores += 1

                i += 1

            print("\n=== ESTADÍSTICAS ===")
            print("Cantidad de alumnos masculinos:", total_m)
            print("Cantidad de alumnas femeninas:", total_f)
            print("Cantidad de alumnos con lentes:", lentes_m)
            print("Cantidad de alumnas con lentes:", lentes_f)
            print("Cantidad de alumnos y alumnas SIN lentes:", sin_lentes)
            print("Cantidad de alumnos menores de edad:", menores)
            print("Cantidad de alumnos mayores de edad:", mayores)
            print("Sección ingresada:", seccion)
            print("Monto Total recaudado: S/.", total_donado)
            print("Monto luego de descontar refrigerio (S/.50): S/.", total_donado - 50)
            print("Monto Neto Total recaudado: S/.", total_donado - 50)
            print("===========================\n")

    elif menu == 5:
        print("ESTÁ SALIENDO DE TECNICAS DE PROGRAMACION :P")
        break

    else:
        print("Opción no válida. Intente de nuevo.")
```
